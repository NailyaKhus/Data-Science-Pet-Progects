# 2. Подготовка списков марок и моделей

```python
# Загрузим файл с перечнем марок автомобилей
with open('dict_mark.json', 'r', encoding='utf-8') as f: 
    dict_mark = json.load(f) 
```

```python
dict_mark
```

```python
# Удалим не популярные модели с количеством объявлений ниже 300
keys_to_delete = []
for mark in dict_mark.keys():
    if int(dict_mark[mark]) < 300:
        keys_to_delete.append(mark)
    
for key in keys_to_delete: del dict_mark[key]
```

```python
# Удалим модели с русскими названиями
for key in ['LADA (ВАЗ)','ГАЗ','УАЗ']: del dict_mark[key]
```

```python
# Приведем к единому регистру
for mark in list(dict_mark.keys()):
    mark_up = mark.upper()
    dict_mark[mark_up]=dict_mark.pop(mark)
```

```python
# При загрузке в адресной строке сложные названия марок представляются иначе. Приведем к единому виду
dict_mark['MERCEDES']=dict_mark.pop('MERCEDES-BENZ')
dict_mark['LAND_ROVER']=dict_mark.pop('LAND ROVER')
dict_mark['GREAT_WALL']=dict_mark.pop('GREAT WALL')
dict_mark['SSANG_YONG']=dict_mark.pop('SSANGYONG')
```

```python
# Загрузим test и посмотрим какие столбцы нам нужны 
test = pd.read_csv('test.csv')
print('Кол-во столбцов теста:', test.shape[1])
print('Названия столбцов теста:', list(test.columns))

# вывод всех столбцов датафрейма
pd.set_option('display.max_columns', None)
test.sample(2)
```

```python
# Функция для очистки временного словаря
def clean_info_values(car_info):
    for value in car_info:
        car_info[value] = 0
```

```python
# Функция для сбора словаря данных для одной машины
def from_car_info_to_dict(car_in_data, car_info):
    clean_info_values(car_info)

    try:
        car_info['bodyType'] = car_in_data['vehicle_info']['configuration']['human_name']
    except:
        car_info['bodyType'] = None

    try:
        car_info['brand'] = car_in_data['vehicle_info']['mark_info']['code']
    except:
        car_info['brand'] = None

    try:
        car_info['color'] = car_in_data['color_hex']
    except:
        car_info['color'] = None

    try:
        car_info['complectation_dict'] = car_in_data['vehicle_info']['complectation']
    except:
        car_info['complectation_dict'] = None

    try:
        car_info['description'] = car_in_data['description']
    except:
        car_info['description'] = None

    try:
        car_info['engineDisplacement'] = car_in_data['vehicle_info']['tech_param']['human_name'].split()[
            0]
    except:
        car_info['engineDisplacement'] = None

    try:
        car_info['enginePower'] = car_in_data['vehicle_info']['tech_param']['power']
    except:
        car_info['enginePower'] = None

    try:
        car_info['equipment_dict'] = car_in_data['vehicle_info']['equipment']
    except:
        car_info['equipment_dict'] = None

    try:
        car_info['fuelType'] = car_in_data['lk_summary'].split()[-1]
    except:
        car_info['fuelType'] = None

    try:
        car_info['image'] = car_in_data['state']['image_urls'][0]['sizes']['320x240']
    except:
        car_info['image'] = None

    try:
        car_info['mileage'] = car_in_data['state']['mileage']
    except:
        car_info['mileage'] = None

    try:
        car_info['modelDate'] = car_in_data['vehicle_info']['super_gen']['year_from']
    except:
        car_info['modelDate'] = None

    try:
        car_info['model_info'] = car_in_data['vehicle_info']['model_info']
    except:
        car_info['model_info'] = None

    try:
        car_info['model_name'] = car_in_data['vehicle_info']['model_info']['code']
    except:
        car_info['model_name'] = None

    try:
        car_info['name'] = car_in_data['vehicle_info']['tech_param']['human_name']
    except:
        car_info['name'] = None

    try:
        car_info['numberOfDoors'] = car_in_data['vehicle_info']['configuration']['doors_count']
    except:
        car_info['numberOfDoors'] = None

    try:
        car_info['priceCurrency'] = car_in_data['price_info']['currency']
    except:
        car_info['priceCurrency'] = None

    try:
        car_info['productionDate'] = car_in_data['documents']['year']
    except:
        car_info['productionDate'] = None

    try:
        car_info['sell_id'] = car_in_data['id']
    except:
        car_info['sell_id'] = None

    try:
        car_info['super_gen'] = car_in_data['vehicle_info']['tech_param']
    except:
        car_info['super_gen'] = None

    try:
        car_info['vehicleConfiguration'] = car_in_data['vehicle_info']['configuration']['body_type']+' ' + \
            car_in_data['vehicle_info']['tech_param']['transmission']+' ' + \
            car_in_data['vehicle_info']['tech_param']['human_name'].split()[0]
    except:
        car_info['vehicleConfiguration'] = None

    try:
        car_info['vehicleTransmission'] = car_in_data['vehicle_info']['tech_param']['transmission']
    except:
        car_info['vehicleTransmission'] = None

    try:
        car_info['vendor'] = car_in_data['vehicle_info']['vendor']
    except:
        car_info['vendor'] = None

    try:
        car_info['Владельцы'] = car_in_data['documents']['owners_number']
    except:
        car_info['Владельцы'] = None

    try:
        car_info['Владение'] = car_in_data['documents']['purchase_date']
    except:
        car_info['Владение'] = None

    try:
        car_info['ПТС'] = car_in_data['documents']['pts']
    except:
        car_info['ПТС'] = None

    try:
        car_info['Привод'] = car_in_data['lk_summary'].split(', ')[-2]
    except:
        car_info['Привод'] = None

    try:
        car_info['Руль'] = car_in_data['vehicle_info']['steering_wheel']
    except:
        car_info['Руль'] = None

    try:
        car_info['Состояние'] = car_in_data['status'] == 'ACTIVE'
    except:
        car_info['Состояние'] = None

    try:
        car_info['Таможня'] = car_in_data['documents']['custom_cleared']
    except:
        car_info['Таможня'] = None

    try:
        car_info['tax_by_year'] = car_in_data['owner_expenses']['transport_tax']['tax_by_year']
    except:
        car_info['tax_by_year'] = None

    try:
        car_info['transport_tax_rid'] = car_in_data['owner_expenses']['transport_tax']['rid']
    except:
        car_info['transport_tax_rid'] = None

    try:
        car_info['insurance_company'] = car_in_data['owner_expenses']['osago_insurance'][0]['insurance_company']
    except:
        car_info['insurance_company'] = None

    try:
        car_info['osago_insurance_price'] = car_in_data['owner_expenses']['osago_insurance'][0]['price']
    except:
        car_info['osago_insurance_price'] = None

    try:
        car_info['start_date'] = car_in_data['additional_info']['hot_info']['start_time']
    except:
        car_info['start_date'] = None

    try:
        car_info['auto_class'] = car_in_data['vehicle_info']['configuration']['auto_class']
    except:
        car_info['auto_class'] = None

    try:
        car_info['price_RUR'] = car_in_data['price_info']['RUR']
    except:
        car_info['price_RUR'] = None

    try:
        car_info['price_EUR'] = car_in_data['price_info']['EUR']
    except:
        car_info['price_EUR'] = None

    try:
        car_info['price_USD'] = car_in_data['price_info']['USD']
    except:
        car_info['price_USD'] = None

    return car_info
```

# 4. Парсинг

```python
# Словарь для сбора данных по одной машине
colums_dict = {'bodyType': 0, 'brand': 0, 'color': 0, 'complectation_dict': 0,
               'description': 0, 'engineDisplacement': 0, 'enginePower': 0, 
               'equipment_dict': 0, 'fuelType': 0, 'image': 0, 'mileage': 0, 'modelDate': 0,
               'model_info': 0, 'model_name': 0, 'name': 0, 'numberOfDoors': 0, 
               'priceCurrency': 0, 'productionDate': 0, 'sell_id': 0,
               'super_gen': 0, 'vehicleConfiguration': 0, 'vehicleTransmission': 0,
               'vendor': 0, 'Владельцы': 0, 'Владение': 0, 'ПТС': 0, 'Привод': 0, 'Руль': 0,
               'Состояние': 0, 'Таможня': 0, 'tax_by_year': 0, 'transport_tax_rid': 0, 
               'insurance_company': 0, 'osago_insurance_price': 0, 'start_date': 0, 'auto_class': 0, 
               'price_RUR': 0, 'price_EUR': 0, 'price_USD': 0}
```

```python
# url для парсинга auto.ru
url = "https://auto.ru/-/ajax/desktop/listing/"
```

```python
HEADERS = {
    'Accept': '*/*',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Content-Length': '<calculated when request is sent>',
    'content-type': 'application/json',
    'Cookie': '_csrf_token=69292abc547f3cfabac9d3d454b678c67f6471d398451743',
    'Host': 'auto.ru',
    'x-csrf-token': '69292abc547f3cfabac9d3d454b678c67f6471d398451743',
}
```

```python
auto_ru = {}  # итоговый словарь для записи
iteration = 0

for mark in list_of_marks:

    list_of_models = json.loads(
        df_models.loc[mark, 'Models'].replace("'", '"'))

    for model in list_of_models:

        for num_page in range(1, 200):

            param = {
                "catalog_filter": [{"mark": mark, "model": model}],
                "section": "all",
                "category": "cars",
                "sort": "fresh_relevance_1-desc",
                "top_days": 900,
                "geo_radius": 200,
                "geo_id": [213],
                "page": num_page
            }

            # Из-за превышения времени ожидания ответа цикл может обрываться.
            # Исключим прерывания блоком try-except

            attempt = 1    # подсчет количества попыток

            while attempt < 4:

                try:
                    response = requests.post(url, json=param, headers=HEADERS)
                    data = response.json()
                    break

                except:
                    attempt += 1
                    print(
                        f'Марка: {mark}. Модель: {model}. Страница: {num_page}.')
                    print(
                        f'Время ожидания ответа превышено. {attempt}-я попытка')
                    time.sleep(1)

            if data['offers'] != []:
                for car in data['offers']:
                    from_car_info_to_dict(car, colums_dict)
                    auto_ru[iteration] = colums_dict.copy()
                    iteration += 1

            else:
                break
print('Готово!')
```

```python
# Представим собранный словарь в виде датасета
df_auto_ru = pd.DataFrame(auto_ru).T
df_auto_ru.sample(5)
```
